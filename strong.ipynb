```python
import cv2
import sys
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
path_TestImg = 'C:/Users/admin/Desktop/ImageAnalysis/Test_Image/'

path_bird = path_TestImg + 'QuteBird.jpg'
path_panda = path_TestImg + 'panda.jpg'
```

```python
# 색체널 분리



def Color_channel(img,color_range=60,brightness_range = (0,255),chroma_range = (0,255)):
    
    
    # color_range max = 60 색의 범위를 결정할 숫자
    # brightness_range = (0,255) 명도
    # chroma_range = (0,255) 채도


    red_lower = np.array([0-(color_range/2),chroma_range[0],brightness_range[0]])
    red_upper = np.array([0+(color_range/2),chroma_range[1],brightness_range[1]])
    green_lower = np.array([60-(color_range/2),chroma_range[0],brightness_range[0]])
    green_upper = np.array([60+(color_range/2),chroma_range[1],brightness_range[1]])
    blue_lower = np.array([120-(color_range/2),chroma_range[0],brightness_range[0]])
    blue_upper = np.array([120+(color_range/2),chroma_range[1],brightness_range[1]])
    
    
    hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
    red_mask = cv2.inRange(hsv,red_lower,red_upper)
    green_mask = cv2.inRange(hsv,green_lower,green_upper)
    blue_mask = cv2.inRange(hsv,blue_lower,blue_upper)
    
    red = cv2.bitwise_and(img,img,mask=red_mask)
    green = cv2.bitwise_and(img,img, mask = green_mask)
    blue = cv2.bitwise_and(img,img,mask = blue_mask)
    
    return red, green, blue



#가우시안 필터
def GaussianBlur(img,ksize):
    blur = cv2.GaussianBlur(img,ksize=(ksize,ksize),sigmaX=0)
    edged  = cv2.Canny(blur,10,250)
    return edged


# 제곱이미지
def image_square(temp):
    temp = temp**1.9999999999999
    img_s = temp/(np.max(temp)/255)
    return img_s.astype(np.uint8)

# 컬러이미지 히스토그램 평활화
def image_enhanced(temp):
    enhanced_yuv = cv2.cvtColor(temp, cv2.COLOR_BGR2YUV)
    enhanced_yuv[:, :, 0] = cv2.equalizeHist(enhanced_yuv[:, :, 0])
    return cv2.cvtColor(enhanced_yuv, cv2.COLOR_YUV2RGB)

# 흑백이미지 히스토그램 평활화
def gray_enhanced(temp):
    enhanced_yuv = cv2.cvtColor(temp, cv2.COLOR_BGR2YUV)
    enhanced_yuv[:, :, 0] = cv2.equalizeHist(enhanced_yuv[:, :, 0])
    return cv2.cvtColor(enhanced_yuv, cv2.COLOR_YUV2RGB)


def edged_closed(gaussian):
    edged = cv2.Canny(gaussian, 10, 250)
    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5,5))
    return cv2.morphologyEx(edged, cv2.MORPH_CLOSE, kernel)


# 제곱이미지 -> 평활화 -> 가우시안 -> 선정리
def square_closed(img):
    temp = cv2.cvtColor(img.copy(),cv2.COLOR_BGR2RGB)
    
    # 제곱이미지
    square = image_square(temp)
    
    # 평활화
    enhanced =   image_enhanced(square)
    
    # 가우시안
    gaussian = GaussianBlur(enhanced,1)

    # 가우시안 선정리
    closed = edged_closed(gaussian)
    
    return enhanced , gaussian , closed



def img_sum(img):
    if 1 < len(img):
        result = img[0]
        for i in img[1:]:
            result = cv2.add(result,i)
        return result
    else:
        return img
```

```python
origin_bird = cv2.imread(path_bird)

hsv = Color_channel(origin_bird)


f,ax = plt.subplots(3,4)
f.set_size_inches(15,11)
ax[0,0].imshow(origin_bird)
ax[0,1].imshow(hsv[0])
ax[0,2].imshow(hsv[1])
ax[0,3].imshow(hsv[2])

ax[1,0].imshow(square_closed(origin_bird)[1])
ax[1,1].imshow(square_closed(hsv[0])[1])
ax[1,2].imshow(square_closed(hsv[1])[1])
ax[1,3].imshow(square_closed(hsv[2])[1])


ax[2,0].imshow(square_closed(origin_bird)[2])
ax[2,1].imshow(square_closed(hsv[0])[2])
ax[2,2].imshow(square_closed(hsv[1])[2])
ax[2,3].imshow(square_closed(hsv[2])[2])
```

```python
bird_hsv= Color_channel(origin_bird)

# temp = cv2.add(square_closed(origin_bird)[1],GaussianBlur(bird_hsv[0],1)) 원본도 더함

temp = square_closed(bird_hsv[0])[1] # 원본은 안더함
temp = cv2.add(temp,square_closed(bird_hsv[1])[1])
temp = cv2.add(temp,square_closed(bird_hsv[2])[1])

sum_gaussian_bird = edged_closed(
    img_sum([temp,square_closed(bird_hsv[0])[1],
             temp,square_closed(bird_hsv[1])[1],
             temp,square_closed(bird_hsv[2])[1]]))

plt.imshow(sum_gaussian_bird)
```

```python
bird_hsv= Color_channel(origin_bird)

bird_red = bird_hsv[0]
bird_green = bird_hsv[1]
bird_blue = bird_hsv[2]

bird_red_sc = square_closed(cv2.medianBlur(bird_red,7))[1]
bird_green_sc = square_closed(cv2.medianBlur(bird_green,7))[1]
bird_blue_sc = square_closed(cv2.medianBlur(bird_blue,7))[1]


f,ax = plt.subplots(1,4)
f.set_size_inches(15,11)
ax[0].imshow(img_sum([bird_red_sc,bird_green_sc,bird_blue_sc]))
ax[1].imshow(bird_red_sc)
ax[2].imshow(bird_green_sc)
ax[3].imshow(bird_blue_sc)
```
