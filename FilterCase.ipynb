```python
import cv2
import sys
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('C:/Users/admin/Downloads/QuteBird.jpg')
img_gray = cv2.imread('C:/Users/admin/Downloads/QuteBird.jpg',cv2.IMREAD_GRAYSCALE)
b,g,r = cv2.split(img)
img_2 = cv2.merge([r,g,b])
display(plt.imshow(img_2))
```

```python
cv2.imshow('img_gray',img_gray)
cv2.imshow('img_origin',img_2)
cv2.waitKey(0)
```

```python
blur_GaussianBlur = cv2.GaussianBlur(img_gray,ksize=(5,5),sigmaX=0)
edged_GaussianBlur  = cv2.Canny(blur_GaussianBlur,10,250)
display(plt.imshow(edged_GaussianBlur))
```

```python
blur_bilateralFilter = cv2.bilateralFilter(img,-1,80,80)
display(plt.imshow(blur_bilateralFilter))
```

```python
cv2.imshow('blur_bilateralFilter',blur_bilateralFilter)
cv2.imshow('img_origin',img_2)
cv2.waitKey(0)
```
